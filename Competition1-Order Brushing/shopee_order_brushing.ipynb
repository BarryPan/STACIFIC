```python
#取同一天
df['date']=df['event_time'].apply(lambda x:x[8:10]) 
#group by & 取時間最大
df_time_max=df.groupby(['userid','shopid','date']).agg({'event_time':max})  
#group by & 取時間最小
df_time_min=df.groupby(['userid','shopid','date']).agg({'event_time':min})  
```

```python
df_time_max1=df_time_max.reset_index()  #移除多層level
df_time_min1=df_time_min.reset_index()
```

```python
import datetime as dt
#合併
df_time_merge=df_time_min.merge(df_time_max1,on=['userid','shopid','date']).rename(columns={'event_time_x':'min_time','event_time_y':'max_time'}) 
#篩選最早和最晚不同時間資料
df_time_merge_diff=df_time_merge[df_time_merge['min_time'] != df_time_merge['max_time']] 

df_gp_date=df.groupby(['userid','shopid','date']).count()
df_group_date_time=df_gp_date.merge(df_time_merge_diff,on=['userid','shopid','date']) 
```

```python
from datetime import datetime
#轉換時間string to datetime
df_group_date_time['min_time']=df_group_date_time['min_time'].apply(lambda x:datetime.strptime(x,'%Y-%m-%d %H:%M:%S'))
df_group_date_time['max_time']=df_group_date_time['max_time'].apply(lambda x:datetime.strptime(x,'%Y-%m-%d %H:%M:%S'))
#計算時間差距
df_group_date_time['delta_time']=df_group_date_time['max_time']-df_group_date_time['min_time']   
```

```python
import numpy as np
#取時間差距計算為分鐘
df_group_date_time['delta_time']=(df_group_date_time['max_time']-df_group_date_time['min_time'])/np.timedelta64(1, 'm')
#計算每次交易平均過多少分鐘
df_group_date_time['average_time']=df_group_date_time['delta_time']/df_group_date_time['orderid']
#取useid在同shopid 交易次數>2
df_group_date_time_sup=df_group_date_time[df_group_date_time['orderid']>2]
#交易間隔時間中位數為6.6，取交易間隔時間過短6分鐘以下
df_group_date_time_sup=df_group_date_time_sup[df_group_date_time_sup['average_time']<6]
df_group_date_time_sup.head()
```
